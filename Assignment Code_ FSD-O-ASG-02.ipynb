# Question - 1

Question 1: You just joined a company as a junior developer. Your team lead asks you to go to the project folder called myapp, then go inside a subfolder called src, and list all the files there. You open the terminal but you don't know where you are. What commands will you run, step by step?

Answer 1:
```
pwd
ls
cd myapp
cd src
ls
```

**Explanation:**

- ```pwd``` - Shows your current directory (so we know where we are)
- ```ls``` - Lists files/folders in the current directory
- ```cd myapp``` - Moves into the myapp folder
- ```cd src``` - Moves into the src folder inside myapp
- ```ls``` - Lists all files inside src

# Question - 6

Question 6: You are running a command that generates a long list of files. You want to see only the files whose names contain the word 'config'. How will you combine two commands to do this in one line?

Answer 6:


```
ls | grep "config"
```

**Explanation:**

- ```ls``` - lists all files and directories
- ```|``` - (pipe) sends the output of ls to another command
- ```grep "config"``` - filters and shows only those files whose names contain the word "config"

The pipe operator ```|``` is used to combine commands by passing the output of one command to another, allowing filtered results.



# Question - 7

Question 7: A company's website is not working. The developer checks the HTML file and sees there is no DOCTYPE declaration at the top and the content is placed directly without any head or body tags. Why could this be a problem?

Answer 7:  

If an HTML file does not include a ```DOCTYPE``` declaration and proper structure like ```<head>``` and ```<body>``` tags, it can cause several issues

- Missing ```DOCTYPE``` -
The browser may not know which version of HTML is being used. This can lead to inconsistent or incorrect rendering.

- Missing ```<head>``` tag -
Important elements like metadata, title, CSS links, and scripts cannot be properly definedm, which can affect SEO and page behavior.

- Missing ```<body>``` tag -
The main content of the webpage is not properly structured, which can cause display and layout issues.

Proper HTML structure including DOCTYPE, ```<head>```, and ```<body>``` ensures correct rendering, better compatibility across browsers, and proper organization of web content.

# Question - 9

Question 9: What is the difference between > and >> in the terminal? Give one example of when you would use each.

Answer 9:  
- ```>``` → Overwrites the content of a file.
- ```>>``` → Appends (adds) content to the end of a file.

**Example -**


```
echo "Hello" > file.txt
```
  - If ```file.txt``` already has content, it will be deleted and replaced with ```"Hello"```


```
echo "Hello" >> file.txt
```
  - ```"Hello"``` will be added at the end of existing content without deleting it.


**The ```>``` operator replaces the existing content of a file, while ```>>``` adds new content to the file without removing the existing data.**





# Question - 10

Question 10: What is the difference between rm and rm -r? Why should you be careful when using rm -r?

Answer 10:
- rm - Deletes files only.
- rm -r - Deletes directories and their contents recursively.

---

**Example -**

``` rm file.txt ```
  - Deletes a single file.

``` rm -r foldername ```
  - Deletes the folder and everything inside it (files + subfolders)

---

Why we should be careful -

- ```rm -r``` can permanently delete large amounts of data
- There is no recycle bin / undo
- A wrong command can delete important system files
- The ```rm -r``` command removes directories recursively, so it must be used carefully as it can permanently delete all files and folders inside a directory.

---

```rm -ri foldername```
  - Asks for confirmation before deleting each file

# Question - 11

Question 11: What is the difference between semantic HTML tags like section and article versus a generic div tag? Why does it matter?

Answer 11:

- ```<section>``` Used to define a thematic grouping of content, such as chapters, features, or different parts of a webpage.


- ```<article>``` Used for independent, self-contained content like blog posts, news articles, or product cards that can stand alone.


- ```<div>``` A non-semantic container used mainly for styling or layout purposes. It does not convey any information about the type or importance of the content inside it.


**Why it matters ?**

- **Better Readability** - Developers can easily understand the structure of the webpage

- **Improved SEO** - Search engines can better index and rank the content

- **Accessibility** - Screen readers can interpret the page more effectively for visually impaired users

- **Maintainability** - Easier to manage and update code in large projects

**Conclusion -**

Semantic HTML tags provide meaningful structure and context to web content, improving readability, SEO, accessibility, and maintainability, while ```<div>``` is a generic container used only for layout without any semantic meaning.

# Question - 12

Question 12: Why are forms important in web development? Name three input types used in HTML forms and explain what each is used for.

Answer 12:

Forms are an essential part of web development because they allow users to enter and submit data to a website, such as login details, contact information, feedback, or orders.

**Why forms are important:**
- Enable user interaction with websites
- Collect user data (e.g., name, email, password)
- Used in login systems, registrations, payments, and contact pages


**Three input types with explanation:**

- ```text``` Used to take short text input from users (e.g., name, username)

- ```email``` Used to take email addresses and provides built-in validation (must include @)

- ```password``` Used to take sensitive data, characters are hidden for security

**Example:**
```
<form>
  <input type="text" placeholder="Enter your name">
  <input type="email" placeholder="Enter your email">
  <input type="password" placeholder="Enter your password">
</form>
```

**Conclusion -**
Forms are used to collect user input and enable interaction between users and websites. Input types like text, email, and password help gather different kinds of user data efficiently and securely.
