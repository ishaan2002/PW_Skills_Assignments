# Question No 3

**Question 3:** Your cousin is learning to code and wrote a program in Python. She clicked "Run" and got an error saying "SyntaxError". She is confused and asks you — "I wrote everything correctly, why is the computer not understanding me?" Explain to her why computers cannot directly understand what we write, and what happens
between writing code and the computer actually running it.

**Answer 3:** To explain this properly, we need to understand that, Programming languages like Python, Java, or C++ are designed to be human-readable, not directly understandable by computers. When we write code, it must go through several steps before the computer can execute it.

Computers only understand machine language, which is made up of binary numbers (0 and 1). However, writing programs in binary is extremely difficult for humans. So, programming languages act as a bridge between humans and machines.

If you tell a computer - `print(Hello) `

You may think it is correct, but the computer sees:
 - Missing quotation mark
 - Incorrect syntax

So it cannot process the instruction.

**Syntax Error -** A Syntax Error occurs when the rules of the programming language are not followed properly. Just like grammar rules in English, programming languages also have strict rules. If these rules are violated, the computer cannot understand the instructions.

**What Happens Between Writing Code and Running It -**
1. Writing the Code (Source Code)
    - My cousin writes code in Python: `print("Hello World")`
    - This is called source code, which is written in a high-level   programming language.

2. Checking the Syntax
    - Before running the code, Python checks whether the code follows proper rules (syntax).
    - These rules include:
        - Correct spelling of keywords
        - Proper use of brackets and quotes
        - Correct indentation (very important in Python)
    - If any rule is broken, Python throws a SyntaxError.

3. Translation by Interpreter
    - Python uses something called an interpreter.
    - The interpreter reads code line by line
    - Converts it into machine-understandable instructions
    - Executes it immediately
    - If it finds an error, it stops and shows a message.

4. Execution by the Computer
    - If the code is correct, It gets translated into machine language
    - The CPU executes it
    - Output is shown on the screen

**Conclusion -** Syntax Error does not mean the idea is wrong—it simply means the way it is written does not follow the rules of the programming language. Computers require precise and correctly structured instructions, and any small mistake can prevent execution.

Errors are a normal part of learning programming. With practice, attention to detail, and proper debugging, these errors can be easily resolved.

# Question No 4

**Question 4:**
You are asked to write step-by-step nstructions (algorithm) for this problem: "A customer is buying items from an online store. If the total bill is more than ₹500, give a 10% discount. Otherwise, no discount." Write the solution first in plain English steps, and then convert it into simple pseudocode.

**Answer 4:**
**Algorithm in Plain English Steps -**

1. Start the process
2. Take the total bill amount as input from the customer
3. Check if the total bill amount is greater than ₹500
4. If the bill is greater than ₹500:
      - Calculate 10% discount on the total bill
      - Subtract the discount from the total bill
      - Store the final amount
5. Otherwise:
      - No discount is applied
      - Final amount remains the same as total bill
6. Display the final payable amount
7. End the process

**Explanation of the Algorithm -**
 - First, we collect the input (bill amount)
 - Then we apply a condition (greater than ₹500)
 - Based on the condition, we either apply a  discount or not
 - Finally, we display the result

**Pseudocode -**

```
START

INPUT total_bill

IF total_bill > 500 THEN
    discount = total_bill * 0.10
    final_amount = total_bill - discount
ELSE
    final_amount = total_bill
END IF

PRINT final_amount

END
```
**Explanation of Pseudocode -**
- INPUT total_bill → Takes value from user
- IF condition → Checks whether bill is greater than 500
- discount calculation → 10% of total bill
- final_amount → Amount after applying discount
- PRINT → Displays output




# Question No 5

**Question 5:**
Your younger brother opens his phone browser and types "www.flipkart.com". Within 2 seconds, the Flipkart homepage appears on his screen. He asks you — "How did this happen so fast? Where did this page come from?" Explain the complete journey — from the moment he types the website name to the moment the page shows up on his screen. Use simple terms like DNS, server, browser, HTTP/HTTPS in your explanation.

**Answer 5:**

**Introduction -**
The internet works on a system of interconnected computers (called servers) that communicate with each other. When we open a website, our request travels across the internet, reaches the correct server, and then the server sends back the required webpage.

This entire process happens in just a few seconds due to highly optimized systems and technologies.


**Complete journey of Opening a Website -**
1. User Enters Website URL in Browser
    - My brother types: ```www.flipkart.com```
    - The browser (like Chrome, Edge, or Firefox) now needs to find where this website exists on the internet.

2. DNS (Domain Name System)
    - Humans use names like ```flipkart.com```, but computers use IP addresses (like ```123.456.789.000```)
    - So, the browser asks the DNS server: What is the IP address of ```www.flipkart```.com ?
    - DNS works like a phonebook of the internet.
    - It converts domain name into IP address
    
3. Browser Sends Request to Server
    - Once the IP address is found, the browser sends a request to the server where Flipkart is hosted.
    - This request uses:
        - HTTP (HyperText Transfer Protocol) or
        - HTTPS (Secure version of HTTP)

4. Server Processes the Request
    - The server understands the request
    - Fetches required data (HTML, CSS, JavaScript, images)
    - Prepares the response
    - Server Sends Response Back to Browser

6. Browser Renders the Web Page
    - Reads HTML to create structure
    - Applies CSS for styling
    - Executes JavaScript for dynamic behavior
    - Finally, the Flipkart homepage appears on the screen.


**HTTP vs HTTPS -**

- HTTP -> Normal communication (not secure)
- HTTPS -> Secure communication using encryption


**Conclusion -** When a user types a website name, a series of steps take place involving DNS lookup, server communication, and browser rendering. Technologies like HTTP/HTTPS, caching ensure that this entire process happens within seconds.

**Question 7:**
What is the difference between a compiler and an interpreter? Explain in your own words with a simple real-life comparison.

**Answer 7:**
As we know, computers cannot understand programming languages like Python, Java, or C++ directly. They only understand machine language (binary). Therefore, we need special tools that translate our code into a form that the computer can execute.

These tools are:

1. Compiler
2. Interpreter

**What is a Compiler?**

A compiler is a program that converts the entire source code at once into machine code before execution. If there are errors, the compiler shows them all together and stops execution until they are fixed.

**Working of a compiller -**

  - Takes the complete program as input
  - Checks for errors
  - Converts it into machine code
  - Generates an executable file

**Languages Using Compiler -**
  - C
  - C++
  - Go

What is an Interpreter?

An interpreter translates and executes code line by line. If an error occurs, it stops at that line and shows the error instantly.

**Working of a Interpreter -**
  - Reads one line of code
  - Translates it
  - Executes it immediately
  - Moves to next line

**Languages Using Interpreter -**
  - Python
  - JavaScript

**Real-Life Comparison: (Cooking method)**

*Compiler = Preparing Full Meal After Reading Recipe*

Imagine you have a recipe book. You read the entire recipe first. Then you prepare the whole dish at once. After cooking, you serve the complete meal

If there is a mistake in the recipe, You realize it only after finishing the cooking

This is like a compiler:
  - Reads full code first
  - Then executes
  - Errors are found after full checking


*Interpreter = Cooking Step-by-Step While Reading*

Now imagine you: Read one step at a time. Perform that step immediately. Then move to the next step

If a step is wrong: You notice it immediately and stop

This is like an interpreter:
  - Executes line by line
  - Stops as soon as it finds an error

**Summary -**

Compiler: Read everything -> then do everything
Interpreter: Read one step -> do one step


**Conclusion -** Both compiler and interpreter are essential tools that help convert programming code into machine language. The main difference lies in how they process the code—compilers translate the entire program at once, while interpreters execute it line by line.

Understanding this difference helps programmers choose the right language and approach based on their needs. Both methods play a crucial role in modern software development, and each has its own advantages depending on the situation.

# Question No 8

**Question 8:**
What is the Software Development Life Cycle (SDLC)? Name all its phases and write one line about what happens in each phase.

**Answer 8:**
The Software Development Life Cycle (SDLC) is a structured process followed by developers to design, develop, test, and maintain software in an organized and efficient way. It acts like a roadmap that guides the development team from the initial idea to the final product.

In real-world software development, building an application without a proper process can lead to confusion, errors, delays, and poor quality. SDLC ensures that software is developed in a systematic, predictable, and high-quality manner.

**Phases of SDLC -**

The SDLC consists of several important phases. Each phase plays a crucial role in building reliable and efficient software.

1. Requirement Analysis - In this phase, the development team collects and understands what the user or client actually needs.
    - Discuss requirements with stakeholders
    - Identify features and functionalities
    - Prepare requirement documents

2. System Design - In this phase, the overall structure of the software is planned.
    - Design system architecture
    - Define database structure
    - Decide technologies to be used

3. Implementation (Coding) - This is the phase where actual coding takes place.
    - Developers write code
    - Convert design into working software
    - Use programming languages and tools

4. Testing - After coding, the software is tested to find and fix errors.
    - Perform manual and automated testing
    - Identify bugs
    - Ensure software works correctly

5. Deployment - In this phase, the software is released to users.

    - Install software on servers or systems
    - Make it available to users
    - Ensure smooth launch

6. Maintenance - After deployment, the software is maintained and improved.
    - Fix bugs
    - Add new features
    - Improve performance

**Flow of SDLC -**

The phases usually follow this order:

1. Requirement Analysis
2. Design
3. Implementation
4. Testing
5. Deployment
6. Maintenance

Each phase depends on the previous one, ensuring a smooth and organized workflow.


**Importance of SDLC -**

Importance of SDLC are as follows :-
- Provides a clear structure
- Reduces errors and confusion
- Improves quality of software
- Helps in proper planning and management
- Ensures timely delivery

**Conclusion -** Software Development Life Cycle (SDLC) is an essential framework that guides the entire process of software development. It ensures that every stage—from understanding requirements to maintaining the final product—is handled systematically.

By following SDLC, developers can build high-quality, reliable, and efficient software while minimizing errors and improving productivity. It is a fundamental concept that every software professional must understand and apply in real-world projects.

# Question No 10

**Question 10:**
What is the difference between HTTP and HTTPS? Why is HTTPS important when you are using a banking or payment website?

**Answer 10:**
When we browse websites, especially for banking or online payments, we often notice two types of URL's:
1. HTTP (HyperText Transfer Protocol)
2. HTTPS (HyperText Transfer Protocol Secure)

They may look similar, but they differ significantly in terms of security and data protection. Understanding this difference is very important, especially in today's digital world where sensitive information like passwords, OTPs, and bank details are frequently shared online.

**What is HTTP?**

HTTP stands for HyperText Transfer Protocol. It is the basic protocol used for communication between a web browser and a server.

How HTTP Works:
- The browser sends a request to the server
- The server sends back the response (webpage data)
- Data is transferred in plain text
- Key Point : HTTP is not secure because data is not encrypted.

**What is HTTPS?**

HTTPS stands for HyperText Transfer Protocol Secure. It is the secure version of HTTP.

How HTTPS Works:
- Uses SSL/TLS encryption
- Encrypts data before sending it over the internet
- Only the intended receiver can decrypt and read the data
- Key Point: HTTPS ensures secure and encrypted communication

**Understanding with a Simple Real-Life Example -**

- Imagine sending a letter:
    - HTTP = Sending a postcard
        - Anyone can read it during transit
        - No privacy
    - HTTPS = Sending a sealed envelope
        - Only the receiver can open it
        - Information is secure

**Real-World Scenario -**

Suppose you are making a payment on an e-commerce website:

- With HTTP:
    - Card details are sent as plain text
    - Hacker can steal information
- With HTTPS:
    - Data is encrypted
    - Even if intercepted, it is unreadable


**Why HTTPS is Important for Banking and Payments -**

When you use banking or payment websites, you share highly sensitive data such as:

- Passwords
- Debit/Credit card details
- OTPs
- Personal information

If HTTP is used:

- Hackers can intercept data
- Information can be stolen

**Conclusion -**
The main difference between HTTP and HTTPS lies in security. HTTP transfers data in plain text and is vulnerable to attacks, while HTTPS encrypts data and ensures safe communication.

HTTPS is extremely important for banking and payment websites because it protects sensitive user information from hackers and ensures data privacy, integrity, and authenticity.

In today's digital world, using HTTPS is not just recommended—it is essential for safe and secure online transactions.
